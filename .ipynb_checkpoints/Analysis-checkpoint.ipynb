# Amazon Analytics

## Imports

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Read in files

```python
df = pd.read_csv('amazon.csv')
```

* product_id - Product ID
* product_name - Name of the Product
* category - Category of the Product
* discounted_price - Discounted Price of the Product
* actual_price - Actual Price of the Product
* discount_percentage - Percentage of Discount for the Product
* rating - Rating of the Product
* rating_count - Number of people who voted for the Amazon rating
* about_product - Description about the Product
* user_id - ID of the user who wrote review for the Product
* user_name - Name of the user who wrote review for the Product
* review_id - ID of the user review
* review_title - Short review
* review_content - Long review
* img_link - Image Link of the Product
* product_link - Official Website Link of the Product


## Data Cleaning

```python
# remove rupees sign from columns with currency
df['discounted_price'] = df['discounted_price'].str.strip('₹')
df['actual_price'] = df['actual_price'].str.strip('₹')
```

```python
df['actual_price'][df['actual_price'].str.split(',').str.len() > 2]
```

```python
df['discounted_price'][df['discounted_price'].str.split(',').str.len() > 2]
```

```python
df = df.drop(249)
```

```python
df['discounted_price'] = df['discounted_price'].str.replace(',', '.')
df['actual_price'] = df['actual_price'].str.replace(',', '.')
```

```python
df['rating_count'] = df['rating_count'].str.replace(',', '')
```

```python
df['discounted_price'][df['discounted_price'].str.split('.').str.len() > 2]
```

```python
df['actual_price'][df['actual_price'].str.split('.').str.len() > 2]
```

```python
df = df.drop([1205, 1352, 1413, 1439, 1455, 873])
```

```python
df['discounted_price'][df['discounted_price'].str.split('.').str.len() > 2]
```

```python
df['actual_price'][df['actual_price'].str.split('.').str.len() > 2]
```

```python
df['discount_percentage'] = df['discount_percentage'].str.strip('%')
```

```python
df[df['rating'] == '|']
```

```python
df = df.drop(1279)
```

```python
# change column types to int
numeric_columns = ['discounted_price', 'actual_price', 'discount_percentage', 
                   'rating', 'rating_count' ]

for col in numeric_columns:
    df[col] = df[col].astype('float')
```

## Analysis

```python
df.info()
```

```python
df.describe()
```

```python
rating_size = df.groupby('rating').size()


plt.figure(figsize=(12, 4))
sns.lineplot(x=rating_size.index, y=rating_size)
plt.title('Individual rating counts')
plt.show()
```

```python
rating_count_top_10 = df.sort_values(by='rating_count', ascending=False).head(11)

plt.figure(figsize=(8, 4))
sns.barplot(data=rating_count_top_10, y='product_name', x='rating_count', orient='h')
plt.title('Top 10 items ranked by "rating_count"')
plt.show()
```

```python
top_10_categories_by_rating = df[['rating', 'category']].groupby('category').mean()['rating'].sort_values(ascending=False).head(10)

plt.figure(figsize=(8, 4))
sns.barplot(y=top_10_categories_by_rating.index, x=top_10_categories_by_rating, orient='h')
plt.title('top 10 categories by rating')
```
